## Características principais:
- Date: A data dos dados de preço das ações.

- Low: O menor preço que a ação atingiu durante o pregão.
- Close: O preço de fechamento da ação nessa data.
- Volume: O volume de negociação, ou seja, o número de ações negociadas naquela data.
- Dividends: Dividendos pagos nessa data (se houver).
- Stock Splits: Informações sobre desdobramentos de ações (se houver).
- Brand_Name: O nome da marca ou empresa.
- Ticker: Símbolo de ticker para a ação.
- Industry_Tag: A ca
egoria de indústria ou setor ao qual a marca pertence.
- Country: O país onde a marca está sediada ou opera principalmente.

## O boxplot
- Permite visualizar os percentis e valores máximo e mínimo dos dados
- Permite visualizar, além de simetria e dispersão dos dados, valores extremos (outliers)
- O valor máximo e mínimo são calculados baseado no interquartil

```python
# Verificando para a coluna Volume

plt.figure(figsize = (10, 6))
sns.boxplot(data = df, x = 'Volume', color = 'blue')
plt.title('Distribuição de Valores da Coluna Volume')
plt.xlabel('Valor')
plt.grid(True)
plt.show()
```

```python
# Criando o boxplot para as colunas Low e High 

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Low', 'High']], orient='v', palette='Set2')
plt.title('Boxplot de Preço Altos e Baixos')
plt.ylabel('Preço')
plt.grid(True)
plt.show()
```

```python
# Determinando o interquartil

volume = df['Volume']

#calculando o primeiro quartil
q1 = volume.quantile(0.25)


#calculando o terceiro quartil
q3 = volume.quantile(0.75)

#calculando o iqr
iqr = q3 - q1

iqr
```

```python
# Calculando o valor máximo e mínimo
volume = df['Volume']

#calculando o primeiro quartil
q1 = volume.quantile(0.25)


#calculando o terceiro quartil
q3 = volume.quantile(0.75)

#calculando o iqr
iqr = q3 - q1

#calculando o limite inferior do intervalo de confiança (LIMITE INFERIOR)
min_value = q1 - 1.5 * iqr

#calculando o limite superior do intervalo de confiança (LIMITE SUPERIOR)
max_value = q3 + 1.5 * iqr 

print(f'Valor Mínimo: {min_value}')
print(f'Valor Máximo: {max_value}')
```

```python
# Verificando a quantidade de registros em cada intervalo (<Q1, entre Q1 e Q2, entre Q2 e Q3,>Q3)

volume = df['Volume']

#calculando o primeiro quartil
q1 = volume.quantile(0.25)


#calculando o terceiro quartil
q3 = volume.quantile(0.75)

# Definindo os limites dos intervalos
inf_limit = q1
sup_limit = q3

# Filtrando os registros em cada intervalo
register_inf_q1 = df[df['Volume'] < inf_limit]
register_bet_q1_q3 = df[(df['Volume'] >= inf_limit) & (df['Volume'] <= sup_limit)]
register_sup_q3 = df[df['Volume'] > sup_limit]

# Contando os registros em cada intervalo
amount_inf_q1 = len(register_inf_q1)
amount_bet_q1_q3 = len(register_bet_q1_q3)
amount_sup_to_q3 = len(register_sup_q3)

print(f'Quantidade de registros abaixo de Q1: {amount_inf_q1}')
print(f'Quantidade de registros no intervalo [Q1, Q3]: {amount_bet_q1_q3}')
print(f'Quantidade de registros acima de Q3: {amount_sup_to_q3}')
```

```python
# Verificando para as colunas Open e Close
open_prices = df['Open']
close_prices = df['Close']

# Calculando os quartis para 'Open'
q1_open = open_prices.quantile(0.25)
q3_open = open_prices.quantile(0.75)

# Calculando os quartis para 'Close'
q1_close = close_prices.quantile(0.25)
q3_close = close_prices.quantile(0.75)

# Definindo os limites dos intervalos para 'Open'
inf_limit_open = q1_open
sup_limit_open = q3_open

# Filtrando os registros em cada intervalo para 'Open'
inf_limit_close = q1_close
sup_limit_close = q3_close

# Filtrando os registros em cada intervalo para 'Open'
register_inf_q1_open = df[df['Open'] < inf_limit_open]
register_bet_q1_q3_open = df[(df['Open'] >= inf_limit_open) & (df['Open'] < sup_limit_open)]
register_sup_q3_open = df[df['Open'] > sup_limit_open]

# Filtrando os registros em cada intervalo para 'Close'
register_inf_q1_close = df[df['Close'] < inf_limit_close]
register_bet_q1_q3_close = df[(df['Close'] >= inf_limit_close) & (df['Close'] < sup_limit_close)]
register_sup_q3_close = df[df['Close'] > sup_limit_close]

# Contando os registros em cada intervalo para 'Open'
amount_inf_q1_open = len(register_inf_q1_open)
amount_bet_q1_q3_open = len(register_bet_q1_q3_open)
amount_sup_q3_open = len(register_sup_q3_open)

# Contando os registros em cada intervalo para 'Close'
amount_inf_q1_close = len(register_inf_q1_close)
amount_bet_q1_q3_close = len(register_bet_q1_q3_close)
amount_sup_q3_close = len(register_sup_q3_close)

print(f'Quantidade de registros abaixo de Q1 para Open: {amount_inf_q1_open}')
print(f'Quantidade de registros entre Q1 e Q3 para Open: {amount_bet_q1_q3_open}')
print(f'Quantidade de registros acima de Q3 para Open: {amount_sup_q3_open}')

print(f'Quantidade de registros abaixo de Q1 para Close: {amount_inf_q1_close}')
print(f'Quantidade de registros entre Q1 e Q3 para Close: {amount_bet_q1_q3_close}')
print(f'Quantidade de registros acima de Q3 para Close: {amount_sup_q3_close}')

```

## Criando uma matriz para mostrar a correlação de cada par de variáveis**
- Podemos usar do pandas o `.plotting.scatter_matrix()` ou do seaborn o `.pairplot()`

```python
# Apresentar a matriz de correção da base

#calculando a matriz de correlação

numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits']

# Criando um DataFrame apenas com as colunas numéricas
numeric_df = df[numeric_columns]

# Calculando a matriz de correlação para cada par de variáveis
correlation_matrices = {}

for column1 in numeric_columns:
    for column2 in numeric_columns:
        if column1 != column2:
            correlation_matrix = numeric_df[[column1, column2]].corr()
            correlation_matrices[f'{column1} vs {column2}'] = correlation_matrix

# Exibindo as matrizes de correlação para cada par de variáveis
for pair, matrix in correlation_matrices.items():
    print(f'Matriz de Correlação para {pair}:\n{matrix}\n')
```

```python
# Apresentar a matriz de correlação da base plotando no pairplot

numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits']

# Criando um DataFrame apenas com as colunas numéricas
numeric_df = df[numeric_columns]

sns.set(style="ticks")
sns.pairplot(numeric_df, kind="scatter")
plt.show()
```

## Visualizando apenas o KDE**
- KDE: Kernel Density Estimation
    - Mede a chance de uma variável aleatória assumir determinado valor
    - A probabilidade é dada pela integral abaixo da curva na faixa de valor selecionada
- Parece uma "versão de linha suavizada" de um histograma

```python
# Fazendo o kde do volume negociado por país 

sns.set(style = 'darkgrid')

# Criando um KDE para o volume negociado por país
sns.kdeplot(data=df, x="Volume", hue="Country", fill=True, common_norm=False, palette="tab10")

# Definindo o título e os rótulos dos eixos
plt.title("KDE do Volume Negociado por País")
plt.xlabel("Volume Negociado")
plt.ylabel("Densidade")
plt.show()
```
